```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter('ignore', FutureWarning)

# Load and normalize data
concrete_data = pd.read_csv('concrete_data.csv')
predictors = concrete_data.drop('Strength', axis=1)
target = concrete_data['Strength']

# Normalize predictors and target
predictors_norm = (predictors - predictors.mean()) / predictors.std()
n_cols = predictors_norm.shape[1]

target_mean = target.mean()
target_std = target.std()
target_norm = (target - target_mean) / target_std

# Split normalized data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(predictors_norm, target_norm, test_size=0.3, random_state=42)

# Build the regression model
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.regularizers import l2
from keras.callbacks import EarlyStopping

def regression_model():
    model = Sequential([
        Dense(10, activation='relu', input_shape=(n_cols,), kernel_regularizer=l2(0.01)),
        Dense(1)
    ])
    
    model.compile(
        optimizer='adam',
        loss='mean_squared_error',
        metrics=['mean_squared_error']
    )
    return model

# Train the model
model = regression_model()
history = model.fit(X_train, y_train, epochs=50, verbose=2)

# Make predictions and denormalize
y_pred = model.predict(X_test) * target.std() + target.mean()
y_test_denorm = y_test * target.std() + target.mean()

# Calculate Mean Squared Error
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test_denorm, y_pred)
print(f"Mean Squared Error (MSE) on Test Data: {mse}")
```
